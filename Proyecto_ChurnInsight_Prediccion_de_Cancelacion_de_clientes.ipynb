La base de datos original escogida es “Customer Churn Dataset" de Muhammad Shahid Azeem. La cual, está disponible en el siguiente enlace: https://www.kaggle.com/datasets/muhammadshahidazeem/customer-churn-dataset
<br><br>
De la cual se extraerán las siguientes columnas para poder crear un modelo predictivo con un mínimo valor viable. El objetivo de este diccionario es comprender la definición de las columnas y sus valores asociados.
Las columnas seleccionadas en un principio son las siguientes:
<br><br>
- **customerID:** Es el código único que identifica al cliente.
- **Tenure:**  Es la permanencia en meses, del cliente dentro de la empresa.
- **payment_delay:** La cantidad de veces que el cliente se ha retrasado con el pago.
- **subscription_type:** Tipo de suscripción del cliente: Basic, Standart, Premium
- **contract_length:** Duración y renovación del contrato: mensual, trimestral, anual.
- **total_spend:** Gasto total del cliente.
- **churn:** Nos muestra si el cliente abandono o no la empresa, con los valores 1, para “si” abandono, y, 0 para “no” abandono
